```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
import pandas as pd

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

def draw_good_average_bad_area_plot(data):
    Cov = data
    columns_titles = ["bad","medium", "good"]
    df = Cov.divide(Cov.sum(axis=1), axis=0)
    df = df.reindex(columns=columns_titles)
    #df.plot.area(legend = 'reverse', xlim = [0,100], ylim = [0,1])
    df.plot.area(legend = False, xlim = [0,100], ylim = [0,1])
#     plt.legend(ncol = 3)
#     legend = plt.legend(ncol = 3)
#     fig  = legend.figure
#     fig.canvas.draw()
#     bbox  = legend.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
#     fig.savefig("../../Pictures/good_bad_legend.pdf", dpi="figure", bbox_inches=bbox)
    plt.xlabel("Time slots")
    plt.gca().xaxis.label.set_size(15)
    plt.gca().yaxis.label.set_size(15)
    plt.tick_params(which='major', labelsize=15)
    #plt.savefig("../../Pictures/"+file_path[:-4]+".pdf")
    plt.show()
    
#draw_good_average_bad_area_plot("n_quality_tasks_RGreedy_M900_1000.csv")
#draw_good_average_bad_area_plot("../csvFilesNorm200steps/n_quality_tasks_fdqo_M900_1000_0.9_baseline0.4_g0.5_queue2k.csv")
```

```python
draw_good_average_bad_area_plot(data1)
draw_good_average_bad_area_plot(data2)
draw_good_average_bad_area_plot(data3)
draw_good_average_bad_area_plot(data4)
draw_good_average_bad_area_plot(data5)
draw_good_average_bad_area_plot(data6)
```
